```python
# Q7 - Display the descriptive statistical analysis on numerical columns of the dataframe?

df.describe()
```

```python
# Q8 - Display the value count on the 'income' column
df.loc[ : ,"income"].value_counts()  # value_counts() function returns the count of unique items in a pandas dataframe.
```

```python
# Q9 - Display the percentage of working men and women in the dataframe?

men = (df.loc[df.gender == "Male","gender"].value_counts().values[0]/df.shape[0])*100 
#we are counting no. of male values and dividing Total no. of column.
print("Total percentage of Men: ",men,"%")

female = (df.loc[df.gender == "Female","gender"].value_counts().values[0]/df.shape[0])*100
print("Total percentage of Female: ",female,"%")
```
